```python
'''Question 1:
Write a program that calculates and prints the value according to the given formula:
Q = Square root of [(2 * C * D)/H]
Following are the fixed values of C and H:
C is 50. H is 30.
D is the variable whose values should be input to your program in a comma-separated
sequence.'''
def cal():
    import math
    l = input('Enter the numbers:')
    d = l.split(',')
    j = []
    for i in d:
        j.append(int(math.sqrt((2 * 50 * int(i)) / 30)))
    print(j)
cal()

'''Question 2:
Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array. The
element value in the i-th row and j-th column of the array should be i*j.
Note: i=0,1.., X-1; j=0,1,Â¡Y-1.'''
def matrix(x,y):
    p = []
    for i in range(x):
        l = []
        for j in range(y):
            l.append(i * j)
        p.append(l)
    return print(p)
matrix(int(input('Enter the number of rows:')),int(input('Enter the number of columns:')))

'''Write a program that accepts a comma separated sequence of words as input and prints
the
words in a
comma - separated
sequence
after
sorting
them
alphabetically.'''

l = (input("Enter the words:"))
b = []
c = ''
for i in l.split(','):
    b.append(i)
b.sort()
for j in range(len(b)):
    c = c + b[j]

    c = c + ','

print(c)


'''Question
4:

Write
a
program
that
accepts
a
sequence
of
whitespace
separated
words as input and prints

the
words
after
removing
all
duplicate
words and sorting
them
alphanumerically.'''

l = (input("Enter the words:  "))
b = []
c = ''
for i in l.split(' '):
    b.append(i)
d = set(b)

b = list(d)

b.sort()

for j in range(len(b)):
    c = c + b[j]

    c = c + ' '

print(c)









'''Question
5:

Write
a
program
that
accepts
a
sentence and calculate
the
number
of
letters and digits.'''
def srchalpha(i):
    p = False

    import re

    pattern = re.compile(r'[a-zA-Z]')

    matches = pattern.finditer(i)

    for match in matches:
        p = True

    return p


def srchnum(i):
    p = False

    import re

    pattern = re.compile(r'[0-9]')

    matches = pattern.finditer(i)

    for match in matches:
        p = True

    return p


d = input('Enter the String :')

LETTERS = 0

DIGITS = 0

for j in list(d):

    if srchalpha(j) == True:
        LETTERS += 1

    if srchnum(j) == True:
        DIGITS += 1

print('LETTERS:', LETTERS)

print('DIGITS:', DIGITS)











'''Question 6: A
website
requires
the
users
to
input
username and password
to
register.Write
a
program
to

check
the
validity
of
password
input
by
users.'''

def passpick(p):
    c = ''

    import re

    pattern = re.compile(r'')

    if re.search(r'[$#@]', p) is not None:

        if re.search(r'[a-z]', p) is not None:

            if re.search(r'[A-Z]', p) is not None:

                if re.search(r'[0-9]', p) is not None:
                    c = c + p

    return c


password = input('Enter sequence of passwords separated by comma: ')

l = password.split(',')

password_list = []

for j in l:

    if len(passpick(j)) > 0 and len(j) >= 6 and len(j) <= 12:
        password_list.append(j)

if len(password_list) > 0:

    print('list of passwords matching criteria :',password_list)

else:

    print('None of the passwords are matching criteria')

```
