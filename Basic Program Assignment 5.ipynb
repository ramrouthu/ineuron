```python
'''1.
Write
a
Python
Program
to
Find
LCM?
'''
n1 = int(input("Enter the number 1 for LCM: "))

n2 = int(input("Enter the number 2 for LCM : "))

count = 0

for i in range(2, (n1 * n2) + 1):

    if i % n1 == 0 and i % n2 == 0:
        print('LCM',i)

        count += 1

        break

if count == 0:
    print('LCM',n1 * n2)

'''2.
Write
a
Python
Program
to
Find
HCF?'''

n = int(input("Enter the number of digits for HCF"))

l = []
hcf=1
for i in range(n):
    l.append(int(input("Enter the Numbers to find HCF :")))

for i in range(max(l), 1, -1):

    count = 0

    for j in l:

        if j % i == 0:
            count += 1

    if count == len(l):
        hcf = i

        break

if count == 0:
    hcf = 1

print("HCF of {} is {}".format(l, hcf))

'''
3.
Write
a
Python
Program
to
Convert
Decimal
to
Binary, Octal and Hexadecimal?
'''

a = int(input("Enter Decimal number to convert :"))

# Decimal to binary Conversion

r = a

e = ""

while r >= 1:
    e = e + str(r % 2)

    r = int(str((r / 2)).split(".")[0])

print("The binary conversion of {} is ".format(a), str(r) + e[::-1])

# Decimal to Octal Conversion

r = a

e = ""

while r >= 7:
    e = e + str(r % 8)

    r = int(str((r / 8)).split(".")[0])

print("The octal conversion of {} is ".format(a), str(r) + e[::-1])

# Decimal to Hexa Conversion

r = a

e = ""

while r >= 15:

    if r == 10 or r % 16 == 10:

        e = e + "A"

    elif r == 11 or r % 16 == 11:

        e = e + "B"

    elif r == 12 or r % 16 == 12:

        e = e + "C"

    elif r == 13 or r % 16 == 13:

        e = e + "D"

    elif r == 14 or r % 16 == 14:

        e = e + "E"

    elif r == 15 or r % 16 == 15:

        e = e + "F"

    else:

        e = e + str(r % 16)

    r = int(str((r / 16)).split(".")[0])

print("The hexa conversion of {} is ".format(a), str(r) + e[::-1])
'''
4.
Write
a
Python
Program
To
Find
ASCII
value
of
a
character?
'''
l = []

for i in range(65, 123):

    if i > 90 and i < 97:
        continue

    l.append(chr(i))

a = input("Enter the Character :")

for i in l:

    if i == a:
        print("The ascii value of {} is {}".format(a, ord(i)))

'''5.
Write
a
Python
Program
to
Make
a
Simple
Calculator
with 4 basic mathematical operations?'''


def add(a, z):
    return print(a + z)


def subtract(a, z):
    return print(a - z)


def multiply(a, z):
    return print(a * z)


def divide(a, z):
    return print(a / z)


print("Enter 1 for Addition")

print("Enter 2 for Subtraction")

print("Enter 3 for Division")

print("Enter 4 for Multiplication")

x = input("Please enter your choice 1/2/3/4 :")

if x == "1":

    m = int(input("Enter number 1: "))

    n = int(input("Enter number 2: "))

    add(m, n)

elif x == "2":

    m = int(input("Enter number 1: "))

    n = int(input("Enter number 2: "))

    subtract(m, n)

elif x == "3":

    m = int(input("Enter number 1: "))

    n = int(input("Enter number 2: "))

    divide(m, n)

elif x == "4":

    m = int(input("Enter number 1: "))

    n = int(input("Enter number 2: "))

    multiply(m, n)

else:

    print("Invalid input")
```
