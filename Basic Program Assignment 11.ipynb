```python
# 1.Write a Python program to find words which are greater than given length k?
p = input('Enter the words separated by comma: ')
k = int(input('Enter the length of words: '))
l = p.split(',')
j = []
for i in l:
    if len(i) >= k:
        j.append(i)
print(j)


# 2. Write a Python program for removing i-th character from a string?

def rem(s, i):
    e = ''
    for j in range(len(s)):
        if j == (i - 1):
            continue
        e += s[j]
    return print(e)


rem('string', 2)


# 3. Write a Python program to split and join a string?

def split_join(s):
    l = []
    for i in s:
        l.append(i)
    print('Splitted String', l)
    s = ''
    for j in l:
        s += j
    print('Joined String', s)


split_join('routhuramganesh')


# 4. Write a Python to check if a given string is binary string or not?
def bin_str(s):
    p = 'Not a binary string'
    import re
    if re.search(r'[^01]', s) is None:
        p = 'Its a binary string'
    return print(p)


bin_str('1')


# 5. Write a Python program to find uncommon words from two Strings?

def str_comp(s1, s2):
    uncommon_words = []
    for i in s1:
        count = 0
        for j in s2:
            if i != j:
                count += 1
        if count == len(s2):
            uncommon_words.append(i)
    for i in s2:
        count = 0
        for j in s1:
            if i != j:
                count += 1
        if count == len(s1):
            uncommon_words.append(i)
    return print(uncommon_words)


str_comp('123', '23g')


# 6. Write a Python to find all duplicate characters in string?
def duplicates(string):
    string_set = set(string)
    duplicate_letters = []
    for i in string_set:
        count = 0
        for j in string:

            if i == j:
                count += 1
        if count > 1:
            duplicate_letters.append(i)
    return print(duplicate_letters)

duplicates('FJHDSGFHJSGDFFSJF')

# 7. Write a Python Program to check if a string contains any special character?
s1 = 'df%ajgd5s'
import re

pattern = re.compile(r'')
if re.search(r'[^a-zA-Z0-9]', s1) is not None:
    print('contains special char')
else:
    print('Donot contain special char')
```
